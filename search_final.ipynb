```python
import numpy as np
import pandas as pd
```

```python
final = pd.read_csv('raw_data/final.csv')
```

```python
def find_course(year, term, department1, course_number1):
    df = final[(final['year'] == year) & 
               (final['term'] == term) &
               (final['course_number1'] == course_number1) &
               (final['department1'] == department1)]
    return df
```
